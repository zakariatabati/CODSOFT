```python
train_df = train_df.loc[:,["unix_time","amt","dob","gender","merchant","category"]]
```

```python
train_df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
train_df['merchant']=encoder.fit_transform(train_df['merchant'])
train_df['category']=encoder.fit_transform(train_df['category'])
train_df['gender']=encoder.fit_transform(train_df['gender'])
train_df['dob']=encoder.fit_transform(train_df['dob'])
```

```python
train_df.head(1)
```

```python
 df = pd.concat([not_fraud,is_fraud])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_df, df["is_fraud"], test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train  = scaler.fit_transform(X_train)
```

```python
X_test  = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import  classification_report,f1_score,confusion_matrix
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(f1_score(y_test,y_pred))
```

```python
 print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.svm import SVC
sv  =SVC()
sv.fit(X_train,y_train)
y_predict = sv.predict(X_test)
print(classification_report(y_test,y_predict))
print("-----------------------------------------")
print(f1_score(y_test,y_predict))
print("-----------------------------------------")
print(confusion_matrix(y_test, y_predict))
```
